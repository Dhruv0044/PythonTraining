```python
# calculator using switch
num=int(input("Enter a number:"))
num2=int(input("Enter the second number:"))
oper=input("Operator")
if (oper=="+"):
    print(num+num2)
elif (oper=="-"):
    print(num-num2)
elif (oper=="*"):
    print(num*num2)
elif (oper=="/"):
    print(num/num2)
elif (oper=="%"):
    print(num%num2)
elif (oper=="**"):
    print(num**num2)
else:
    print("Invalid Operator")
```

```python
#recursion

def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)

factorial(5)
```

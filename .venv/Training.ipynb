```python
# looping through tuple
colors=("red","green","blue","yellow","purple")
for color in colors:
    print(color)
for i in range(len(colors)):
    print(f"{i}: {colors[i]}")
```

Set & Frozen Set

```python
set1={1,2,3,4}
set1=list(set1)
set1=[7,8,9]
set1=set(set1)
print(set1)
set2=set([1,2,2,2,2,2,3,3,3,3,3,3,3,3,30,4,4,4,4,4,4,44,4,4,4])
print(set2)
for item in set2:
    print("Item:",item)
print(set1.symmetric_difference(set2))

```

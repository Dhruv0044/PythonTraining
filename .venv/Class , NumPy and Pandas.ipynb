```python
# Encapsulation ka example h
class bankaccount:
    def __init__(self,salary):
        self.salary=salary
    def  withdraw(self,amount):
        self.salary-=amount
        return self.salary
    def addsalary(self,amount):
        self.salary+=amount
        return self.salary
    def getbalance(self):
        return self.salary
employee1=bankaccount(10000)
print("Balance is :",employee1.getbalance())
print(employee1.addsalary(1000))
print("Balance is :",employee1.getbalance())
print(employee1.withdraw(500))
print("Balance is :",employee1.getbalance())
employee1=bankaccount(1000000000000)
print("Balance is :",employee1.getbalance())
```

```python
#Inheritance ka example
class animal:
    def sound(self):
        print("I am an animal")
class dog(animal):
    def sound(self):
        print("Bark")
c=dog()
c.sound()
```
